```python
def clean_html(html: str) -> str:
    """Remove HTML tags from a string."""
    import re
    clean = re.compile("<.*?>")
    text = re.sub(clean, "", html)
    return re.sub(r"\n{2,}", "\n", text)
```

```python
print(clean_html(html))
```
