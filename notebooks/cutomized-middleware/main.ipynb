```python
from langchain.tools import tool
from langchain.agents import create_agent
```

```python
from src.models import get_model
from src.typings import ModelName
from src.middleware import LoggingMiddleware
```

```python
logging_middleware = LoggingMiddleware(agent="CHIEF_INVESTMENT_OFFICER")
```

```python
langchain_model = get_model("deepseek")
```

```python
from dataclasses import dataclass


@dataclass
class Context:
    runId: str
```

```python
context = Context(runId="123")
```

```python
@tool("example_tool")
def example_tool():
    """Example tool"""
    return "Example tool response"
```

```python
agent = create_agent(
    model=langchain_model,
    middleware=[
        logging_middleware,
    ],
    tools=[example_tool],
    context_schema=Context,
)
```

```python
response = await agent.ainvoke({"messages": [{"role": "user", "content": "please call the example tool multiple times"}]}, context=context)
```
